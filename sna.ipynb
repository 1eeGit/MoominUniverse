```python
dataset = dataset.drop(columns=['Global_ID','File','Group_id','Speaker'])
dataset['speaker_id'], id_index = pd.factorize(dataset['C_Name'])


dataset
```

- remove repeated lines with different timestamp

```python
### ignore all the background slices
dataset = dataset[~dataset['C_Name'].isin(['background_song','background']) ]
### .shift() - compare each row with the previous one
### ~ - NOT
df_cleaned = dataset[~((dataset['Text'] == dataset['Text'].shift())) ]
df_cleaned
```

# Character interactions

## Conversations 
- Adjacent speakers are considered in the same conversation 
- speaker a | b| times they talked

```python
df_conversation = df_cleaned[df_cleaned['speaker_id'] != df_cleaned['speaker_id'].shift()].copy()
df_conversation.loc[:, 'speaker_id'] = df_conversation['speaker_id'].astype('Int64')
```

```python
df_conversation.loc[:, 'prev_speaker_id'] = df_conversation['speaker_id'].shift().astype('Int64')
df_conversation = df_conversation.dropna(subset=['prev_speaker_id'])  # remove first row
```

```python
df = df_conversation.copy()
### removed the group and cmucount
df
```

```python
df_csv = pd.DataFrame()
df_csv['Source'] = df['speaker_id']
df_csv['Target'] = df['prev_speaker_id']
# df_csv['Weight'] = df['transition_count']
df_csv.to_csv('edges.csv', index=False)
```

```python
df_csv2 = pd.DataFrame()
df_csv2['Id'] = df['speaker_id']
df_csv2['Label'] = df['C_Name']
df_csv2.to_csv('nodes.csv', index=False)
```

### additional measurements

```python
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
G = nx.DiGraph()

for _, row in df_csv.iterrows():
    G.add_edge(row['Source'], row['Target'], weight=1)


labels = dict(zip(df_csv2['Id'], df_csv2['Label']))
nx.set_node_attributes(G, labels, "label")
```

```python
### network measurements

nx.set_node_attributes(G, pagerank, "pagerank")
nx.set_node_attributes(G, harmonic, "harmonic")
nx.set_node_attributes(G, katz, "katz")


assortativity = nx.degree_assortativity_coefficient(G)
transitivity = nx.transitivity(G.to_undirected())
efficiency = nx.global_efficiency(G.to_undirected())

print("Assortativity:", assortativity, 'Transitivity:', transitivity, 'Efficiency:', efficiency)
```
